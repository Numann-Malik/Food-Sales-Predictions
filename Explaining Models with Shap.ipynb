```python
import joblib
loaded_joblib = joblib.load('best-models.joblib')
loaded_joblib.keys()
```

```python
X_train_df = loaded_joblib['X_train']
y_train = loaded_joblib['y_train']
X_test_df = loaded_joblib['X_test']
y_test = loaded_joblib['y_test']
preprocessor = loaded_joblib['preprocessor']
rf_model = loaded_joblib['RandomForestRegressor']
linreg_model = loaded_joblib['LinearRegression']
SEED = 42
```

```python
# Import shap and initialize javascript:
import shap
shap.initjs()
```

```python
len(X_train_df)
```

```python
X_shap = shap.sample(X_train_df,nsamples=400,random_state=SEED)
X_shap.head()
```

```python
## get the corresponding y-values
y_shap = y_train.loc[X_shap.index]
y_shap
```

```python
## Resetting the index for sampled data to match shap_values integer index
X_shap = X_shap.reset_index(drop=True)
y_shap = y_shap.reset_index(drop=True)
```

```python
explainer = shap.Explainer(rf_model)
explainer
```

```python
shap_values = explainer(X_shap,y_shap)
shap.summary_plot(shap_values,X_shap)
```

Display our saved feature importance image and compare the most important features according to SHAP vs. your original feature importances.

```python
## Global Force Plot
shap.force_plot(explainer.expected_value, shap_values.values, features=X_shap)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train_df, y_train)
print(f"Training R^2: {lin_reg.score(X_train_df,y_train):.3f}")
print(f"Test R^2: {lin_reg.score(X_test_df, y_test):.3f}")
```

```python
linreg_explainer = shap.Explainer(lin_reg, X_shap)
linreg_explainer
```

```python
linreg_shap_values = linreg_explainer(X_shap)
shap.summary_plot(linreg_shap_values,X_shap)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_reg = KNeighborsRegressor()
knn_reg.fit(X_train_df, y_train)
print(f"Training R^2: {knn_reg.score(X_train_df,y_train):.3f}")
print(f"Test R^2: {knn_reg.score(X_test_df, y_test):.3f}")
```

```python
explainer_knn = shap.Explainer(knn_reg.predict, X_shap)
explainer_knn
```

```python
shap_vals_knn = explainer_knn(X_shap)
shap.summary_plot(shap_vals_knn,X_shap)
```
